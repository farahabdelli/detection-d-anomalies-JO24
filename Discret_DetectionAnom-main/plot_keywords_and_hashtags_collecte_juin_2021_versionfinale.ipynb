```python
import pickle
def load_dataset(dataset_path):
  with open(dataset_path, 'rb') as reader:
      data = pickle.load(reader)


  return (data)

from google.colab import drive
drive.mount('/content/drive')


path1 = "/content/drive/My Drive/ANR-DISCRET/freqtweetjuin.pkl"
df = load_dataset(path1)

path11 = "/content/drive/My Drive/ANR-DISCRET/freqkeywordjuin.pkl"
data20 = load_dataset(path11)

path12 = "/content/drive/My Drive/ANR-DISCRET/freqhashtagsjuin.pkl"
data22 = load_dataset(path12)

path13 = "/content/drive/My Drive/ANR-DISCRET/freqsentimentjuin.pkl"
data23 = load_dataset(path13)

```

```python
#ajouter 2h  aux données
df['created_at'] = pd.to_datetime(df['created_at'].astype(str)) + pd.DateOffset(hours=2)
data20['created_at'] = pd.to_datetime(data20['created_at'].astype(str)) + pd.DateOffset(hours=2)
data22['created_at'] = pd.to_datetime(data22['created_at'].astype(str)) + pd.DateOffset(hours=2)
```

```python
listofkeywords = ["attentat","explosion","accident","incendie","attaque","agression","terroriste","terrorisme","décapitation","bombe","decapitation"," Marseille","Strasbourg","nice","paris","lyon","séisme","seisme","fusillade","TremblementDeTerre","inondation","crue","orage","braquage","otage","secours","ambulance","samu","Iledefrance","Ile_de_france"]
datahashtag = data22[data22['hashtag'].isin(listofkeywords)]
datakeyword = data20[data20['keyword'].isin(listofkeywords)]

```

```python
df['time'] = df['created_at'].dt.round('10min')

aggregation = {'Nombre de tweets': ("count" ,'sum')}
df_freq = df.groupby(['time']).agg(**aggregation).reset_index()


import plotly.express as px

fig = px.line(df_freq, x="time", y="Nombre de tweets")

fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
         
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ))
)


fig.show()
#fig.update_layout(shapes=[shape_dict,shape_dict2,shape_dict3,shape_dict4,shape_dict5,shape_dict6,shape_dict7,shape_dict8,shape_dict9,shape_dict10,shape_dict11,shape_dict12,shape_dict13,shape_dict14,shape_dict15,shape_dict16,shape_dict17,shape_dict18,shape_dict19,shape_dict20,shape_dict21,shape_dict22,shape_dict23,shape_dict24,shape_dict25,shape_dict26,shape_dict27,shape_dict28,shape_dict29,shape_dict30,shape_dict31,shape_dict32,shape_dict33])
fig.write_html("freqjuin.html")
```
